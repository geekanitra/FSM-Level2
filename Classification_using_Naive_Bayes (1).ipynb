```python
def naive_bayes(x_train, y_train):
  features = list(x_train.columns)
  likelihoods = {}
  pred_prior = {}
  class_prior = {}
  class_prior[0]={}
  class_prior[1]={}
  likelihoods["Pclass"] = {}
  likelihoods["Sex"] ={}
  likelihoods["Embarked"] ={}
  pred_prior["Pclass"] = {}
  pred_prior["Sex"] ={}
  pred_prior["Embarked"] ={}

  for feature in features:
    for outcome in np.unique(y_train):
      outcome_count = sum(y_train == outcome)
      feat_likelihood = x_train[feature][y_train[y_train == outcome].index.values.tolist()].value_counts().to_dict()
   
      for feat_val, count in feat_likelihood.items():
        likelihoods[feature][str(feat_val) + '_' + str(outcome)] = count/outcome_count
  
  for feature in features:
			feat_vals = x_train[feature].value_counts().to_dict()

			for feat_val, count in feat_vals.items():
				pred_prior[feature][feat_val] = count/891

  for outcome in np.unique(y_train):
			outcome_count = sum(y_train == outcome)
			class_prior[outcome] = outcome_count /891

  return features, likelihoods, pred_prior, class_prior
```

```python
def predict(class_prior, features, likelihoods, pred_prior, y_test, x_test_modified):

		""" Calculates Posterior probability P(c|x) """

		results = []
		X = np.array(x_test_modified)

		for query in X:
			probs_outcome = {}
			for outcome in np.unique(y_test):
				prior = class_prior[outcome]
				likelihood = 1
				evidence = 1

				for feat, feat_val in zip(features, query):
					likelihood *= likelihoods[feat][str(feat_val) + '_' + str(outcome)]
					evidence *= pred_prior[feat][feat_val]

				posterior = (likelihood * prior) / (evidence)

				probs_outcome[outcome] = posterior

			result = max(probs_outcome, key = lambda x: probs_outcome[x])
			results.append(result)

		return np.array(results)
```

```python
features, likelihoods, pred_prior, class_prior = naive_bayes(x_train, y_train)
```

```python
predictions = predict(class_prior, features, likelihoods, pred_prior, y_test, x_test_modified)
```

```python
def accuracy_score(y_true, y_pred):

	"""	score = (y_true - y_pred) / len(y_true) """

	return round(float(sum(y_pred == y_true))/float(len(y_true)) * 100 ,2)
```
